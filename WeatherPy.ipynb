```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Base URL
base_url = "http://api.openweathermap.org/data/2.5/weather?"

#Create empty lists for each of the data items we want to collect
citylist = []
country = []
date = []
humidity = []
lat = []
lon = []
temp = []
clouds = []
winds = []

#We want to keep track of the cities that we loop through.  In order to do so,
#we count them in records and sets.  Each set has 50 records [1-50].
#We also want to count the number of countries that have to be skipped.
#Set initial recordcount, set count, and skipcount values.
recordcount = 1
setcount = 1
skipcount = 0

#Before beginning data retrieval for each city, print "Beginning Data Retrieval"
print(f"Beginning Data Retrieval\n-----------------------------------")

#loop through each city in our 'cities' list
for each in cities:
    #Set url for each specific city using base url and upi key, and set units to imperial for
    #Fahrenheit readings, and miles per hour in wind speed.
    url = f"{base_url}appid={api_key}&q={each}&units=Imperial"
    data = requests.get(url)
    #If the specific country url cannot be located, print "City not found. Skipping..." and
    #move on to the next city
    if data.status_code == 404:
        print("City not found. Skipping...")
        skipcount+=1
        continue
    #Alternatively if a city is found, print "Processing record," then mention the recordcount
    #number of the set number - and note the city.
    print(f"Processing Record {recordcount} of Set {setcount} | {each}")
    #Add one to the recordcount, for the next city value
    recordcount+=1
    #Keep the recordcount within a set under 50. If we reach 50 in the recordcount, start a new
    #set by seting the recordcount back to one, and adding one to the setcount
    if recordcount==51:
        recordcount = 1
        setcount+=1
    #Put data into json format
    data_json = data.json()
    #Append data that we need for each city to each of our lists.
    citylist.append(data_json['name'])
    clouds.append(data_json['clouds']['all'])
    country.append(data_json['sys']['country'])
    date.append(data_json['dt'])
    humidity.append(data_json['main']['humidity'])
    lat.append(data_json['coord']['lat'])
    lon.append(data_json['coord']['lon'])
    temp.append(data_json['main']['temp_max'])
    winds.append(data_json['wind']['speed'])

print(f"-----------------------------\nWe needed to skip {skipcount} cities.")
print("-----------------------------\nData Retrieval Complete\n-----------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#Create city_df from the data we have collected
city_df = pd.DataFrame({'City': citylist,
    'Cloudiness (%)': clouds,
    'Country': country,
    'Date': date,
    'Humidity (%)': humidity,
    'Latitude': lat,
    'Longitude': lon,
    'Max Temp (F)': temp,
    'Wind Speed (mph)': winds})

#Save city_df output to csv file
city_df.to_csv(output_data_file)

#Display count of all columns in city_df
city_df.count()
```
